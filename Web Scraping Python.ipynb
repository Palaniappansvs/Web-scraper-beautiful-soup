```python
from bs4 import BeautifulSoup
import requests

def scrape(target_url):
    url = target_url
    res = requests.get(url)
    html_page = res.content
    soup = BeautifulSoup(html_page, 'html.parser')
    text = soup.find_all(text=True)
    output = ''
    unwanted_tags = [
    '[document]',
    'noscript',
    'header',
    'html',
    'meta',
    'head', 
    'input',
    'script',
    'div',
    'body',
    'g',
    'svg',
    'style'
    ]

    for t in text:
        if t.parent.name not in unwanted_tags:
            output += '{} '.format(t)

    return output
```

```python
#sample input: http://dsw.accunity.com/
target_url = input('Enter the URL of the webpage: ')
print (scrape(target_url))
```
